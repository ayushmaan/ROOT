```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
data = pd.read_csv('../../DataSet/iris-species/Iris.csv')
data.head()
```

```python
data.shape
```

```python
data['Species'].value_counts()
```

```python
data.plot(kind='Scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
sns.jointplot(data=data,x='SepalLengthCm',y='SepalWidthCm',size=6)
```

```python
sns.FacetGrid(data=data,hue='Species',size=7).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
sns.boxplot(data=data,y='SepalLengthCm',x='Species')
```

```python
ax= sns.boxplot(data=data,y='SepalLengthCm',x='Species')
ax =sns.stripplot(data=data,y='SepalLengthCm',x='Species',jitter=True,edgecolor="gray")
```

```python
sns.violinplot(data=data,y='SepalLengthCm',x='Species')
```

```python
sns.FacetGrid(data=data,size=6,hue='Species').map(sns.kdeplot,'SepalLengthCm').add_legend()
```

```python
sns.pairplot(data=data.drop('Id',1),hue='Species')
```

```python
sns.pairplot(data=data.drop('Id',1),hue='Species',diag_kind='kde')
```

```python
data.drop('Id',1).boxplot(by="Species",figsize=(12,6))
```
